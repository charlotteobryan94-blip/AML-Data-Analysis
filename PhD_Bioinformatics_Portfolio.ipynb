```python
import io
import pandas as pd

# =========================================================================
# LINE-BY-LINE EXPLANATION: GENERATING CANCER PATIENT SAMPLES
# =========================================================================
# csv_header creates a text row containing the standard molecular database columns.
csv_header = "Sample_ID,Gene_Mutated,Mutation_Type,VAF_Percentage,Clinical_Significance\n"

# csv_rows starts a list of text rows representing our first 4 clinical patient profiles.
csv_rows = [
    "TARGET-01,TP53,Missense,42.5,Pathogenic",
    "TARGET-02,FLT3,Internal_Tandem_Duplication,12.1,Pathogenic",
    "TARGET-03,NPM1,Insertion,35.0,Benign_Likely",
    "TARGET-04,KRAS,Missense,51.2,Pathogenic"
]

# This loop automatically generates 196 more patients so we have a full cohort of 200.
for i in range(5, 201):
    # This modulo calculation alternates the simulated mutated genes to keep data realistic.
    gene = "TP53" if i % 3 == 0 else "FLT3" if i % 5 == 1 else "IDH1"

    # This formula creates a varying Variant Allele Frequency (VAF%) between 15% and 60%.
    vaf = 15.0 + (i % 45)

    # This sets the clinical label: if a mutation takes over >30% of a tumor, it is dangerous.
    significance = "Pathogenic" if vaf > 30.0 else "VUS"

    # This formats the data neatly with padded zeros (e.g. TARGET-005) and appends it to our list.
    csv_rows.append(f"TARGET-{i:03d},{gene},Missense,{vaf:.1f},{significance}")

# This joins all 200 patient rows together with a "New Line/Hit Enter" command (\n).
raw_tumor_data = csv_header + "\n".join(csv_rows)

# =========================================================================
# LINE-BY-LINE EXPLANATION: DATA ANALYSIS & FILTERING
# =========================================================================
# io.StringIO converts our raw text string into a virtual file stream for Pandas.
# pd.read_csv reads that stream and builds our interactive spreadsheet named df_biopsy.
df_biopsy = pd.read_csv(io.StringIO(raw_tumor_data))

# high_risk_condition sets our logical filter rule using the AND operator (&).
# It scans for patients who have harmful mutations (Pathogenic) AND a heavy tumor load (VAF > 30%).
high_risk_condition = (df_biopsy['Clinical_Significance'] == 'Pathogenic') & (df_biopsy['VAF_Percentage'] > 30.0)

# The brackets sieve the table, dropping failing rows and keeping the high-risk profiles.
df_high_risk = df_biopsy[high_risk_condition]

# =========================================================================
# LINE-BY-LINE EXPLANATION: PRINTING RELEVENT FINDINGS
# =========================================================================
print("=== CANCER BIOLOGY MUTATION REPORT ===")
print("Total Tumor Biopsies Analyzed:", df_biopsy.shape)
print("High-Risk Clonal Mutations Identified:", df_high_risk.shape)

print("\nFirst 5 High-Risk Patient Profiles:")
print(df_high_risk.head())

# to_csv writes the clean table out of temporary memory into a real file in your Colab sidebar.
df_high_risk.to_csv("high_risk_tumor_clones.csv", index=False)
print("\nFile successfully generated and saved as 'high_risk_tumor_clones.csv'!")
```

```python
import io
import pandas as pd
# ===========================================================================================
# 1. STRUCTURING SEQUENCING MATRIX FOR CELL-LINE EVALUATION
# ===========================================================================================
# We construct a dictionary simulating sequencing data from 5 different CRISPR trials.
# Each record contains the targeting gene locus and the raw cut-site DNA string sequence.

# {...} = Python Dictionary. Use instead of typing raw text streams separated by commas. We define column titles as keys like Cell_Line and map them directly to lists [...] of research records
#pd. DataFrame (...)= takes clean dictrionary key value strucutre and instantly shapes it into an active spreadsheey grid stores inside dr_crispr
crispr_experiment = {"Cell_Line": ["HEK293-T1", "HeLa-KO4", "iPSC-Stem2", "Jurkat-T3", "MCF7-C1"],
    "Target_Gene": ["BRCA1", "PDCD1", "OCT4", "CTLA4", "TP53"],
    "DNA_Sequence": ["GGATCGCCGATC", "ATCGGCTAAGCT", "GCGCGCATATAT", "AAAAATTTTTCC", "GGCGCGATCGCG"],
    "Indel_Efficiency_Pct": [88.5, 42.1, 95.0, 12.4, 76.8]}
# Convert the experiemntal dictionary layout into a form
df_crispr = pd.DataFrame(crispr_experiment)
# ===========================================================================================
# 2. ALGORITHMIC GC-CONTENT DATA CALCULATOR
# ===========================================================================================
# We design a custom bioinformatics function that calculates GC content percentage.
# This metric tells a geneticist how tightly a CRISPR guide binds to its target DNA.
def calculate_gc_percentage(sequence_string):
  #count the instances the Guanine (G) and Cytosine (C) in the string
  g_count = sequence_string.count ('G')
  c_count = sequence_string.count ('C')
  #calculate total percentage baed on total string length
  total_length = len(sequence_string)
  return (( g_count + c_count ) /total_length) * 100
  # We use .apply () to run our custom calculator across every sequence row simultaneously.
  # This appends a brand new calculation column directly to our database table.
df_crispr['GC_Content_Pct'] = df_crispr ['DNA_Sequence'].apply(calculate_gc_percentage)
  # ===========================================================================================
  # 3. CLINICAL FILTERING FOR HIGH- EFFICIENCY TRANSFECTIONS
  # ===========================================================================================
  # In a functional laboratory, cell pool with less than a 70% knockout rate are discarded
  # We slice our database table using a single-bracket filtering to isolate true successes.
df_successful_ko = df_crispr[df_crispr['Indel_Efficiency_Pct']> 70.0]
  # 4. EXPORT AND PRINT CLINICAL RUN REPORT
# =========================================================================
print("=== CRISPR GENE EDITING SCREENING REPORT ===")
print("Total Cell Batches Assayed:\n", df_crispr.shape)
print("\nValidated High-Efficiency Knockouts (>70% Indel):")
print(df_successful_ko[['Cell_Line', 'Target_Gene', 'Indel_Efficiency_Pct', 'GC_Content_Pct']])

# Write the final pipeline results out into a physical workspace file
df_successful_ko.to_csv("successful_crispr_bench.csv", index=False)
print("\nPipeline finished. 'successful_crispr_bench.csv' generated for lab download.")
```
